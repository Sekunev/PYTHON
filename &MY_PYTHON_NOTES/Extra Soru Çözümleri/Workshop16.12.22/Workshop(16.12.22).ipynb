Q 1: Let's write a program that will add the two given lists and create a new list.

2022-06-24 (6).png

```python
list_1 = [1,2,3]
list_2 = [4,5,6]

def addition(x,y):
    newlist = []
    count = 0
    while count < len(x):
        newlist.append(x[count]+y[count])
        count += 1
    return newlist


addition(list_1, list_2)
```

```python
def addition(x,y):
    result = list(map((lambda a,b:a+b), x, y))
    return result

addition(list_1, list_2)
```

Q 2: Let's print the words that are less than 5 letters.
input :  words = ["apple", "swim", "clock", "me", "kiwi", "banana"]
output : swim me kiwi

```python
words = ["apple", "swim", "clock", "me", "kiwi", "banana"]

def filteredlist(iterable):
    result = list(filter((lambda x: len(x) < 5), iterable))
    return result

filteredlist(words)
```

Q4 : input'a girilen index numarasıyla listenin elemanlarına ulaşan kodu yazalım.
İndex numarası yanlış girildiğinde "there is no such an index. Try again!" hatasını,
giriş alfabetik bir karakter olduğunda "you should enter integer. Try again!" hatasını ekrana yazdırsın.
Ayrıca Her hata error.log isimli bir dosyaya tarih ve saatiyle birlikte kaydedilsin

```python
from datetime import datetime
words = ["apple", "swim", "clock", "me", "kiwi", "banana"]

while True :
    try :
        index = int(input("pick an index number :"))
        print("your favorite fruit is", words[index])
        break
    except IndexError as x:
        print("there is no such an index. Try again!")
        with open("error.log", "a") as errorfile:
            errorfile.write(f"{x} errors Time: {datetime.now()}\n")
    except ValueError as y:
        print("you should enter integer. Try again!")
        with open("error.log", "a") as errorfile:
            errorfile.write(f"{y} errors Time: {datetime.now()}\n")


```

```python
def caracters(letters, numbers):
    text = input("enter a text")
    letter, number, space = 0, 0, 0
    for i in text:
        if i in letters:
            # isalpha()
            letter +=1
        elif i in numbers:
            # isnumeric()
            number += 1
        elif i.isspace():
            space += 1

    return (f"string has {letter} letters, {number} numbers and {space} space character.")

caracters(letters, numbers)
```
