During the development of the program, you want the software to have the following features in order to avoid faster and repetitive data entries and to complete the project on time:

when a word is entered into the dictionary to be created;
if the word is not found in the dictionary "This word is not in our database!" give a warning and enable login,
if the word is in the dictionary 'This word is in the database.' to print.

```python
ing_dict = {'blue':'mavi'}
count = 0
while count < 2:
  query = input('Lütfen sorgulayacağınız kelimeyi giriniz. ').lower()
  if query not in ing_dict:
    print('Bu kelime veri tabanında yoktur. Lütfen veri tabanına giriniz.')
    trans = input(f'Lütfen {query} kelimenin anlamını giriniz.').lower()
    ing_dict.update({query : trans})
    count += 1
    print(query, trans)
  else:
    print(query, 'kelimesi veri tabanında bulunmaktadır.')
print('Tebrikler! Bugünkü hedefinizi tamamladınız')
```

Soru 2: Girilen sayıya, karekök'ü tamsayı olan en yakın sayıyı döndüren foksiyonu yazalım. 

image.png

```python
def en_yakın_karekök(sayı):
  karekök = sayı ** (1/2)
  intkarekök = int(karekök)
  sonuç = intkarekök if (abs(intkarekök-karekök) <= abs(intkarekök+1 - karekök)) else intkarekök+1
  return sonuç**2

print(en_yakın_karekök(6))
print(en_yakın_karekök(13))
print(en_yakın_karekök(24)) 
print(en_yakın_karekök(9876))
print(en_yakın_karekök(999999)) 
```
